```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scienceplots

import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
plt.style.use(['science','ieee'])
```

```python
T1 = pd.read_pickle("T_1fids.pickle")
T2 = pd.read_pickle("T_2fids.pickle")
T3 = pd.read_pickle("T_3fids.pickle")
T4 = pd.read_pickle("T_4fids.pickle")
T5 = pd.read_pickle("T_5fids.pickle")
T5_mom = pd.read_pickle("T_5_mome_fids.pickle")
```

```python
T1_keys = list(T1.keys())
T1_values = list(T1.values())
T2_keys = list(T2.keys())
T2_values = list(T2.values())
T3_keys = list(T3.keys())
T3_values = list(T3.values())
T4_keys = list(T4.keys())
T4_values = list(T4.values())
T5_keys = list(T5.keys())
T5_values = list(T5.values())
T5_mom_keys = list(T5_mom.keys())
T5_mom_values = list(T5_mom.values())
```

```python
plt.close()
#plt.figure(figsize = (10,10))
_, ax = plt.subplots(figsize = (5.5,4.5))
y = [10.25]
ax.plot(T1_keys[::3]+T1_keys[-1:],T1_values[::3]+T1_values[-1:],linestyle = '-',color = 'red', marker = 'x', markersize = 1.5)
ax.plot(T2_keys[::4]+T2_keys[-1:],T2_values[::4]+T2_values[-1:],linestyle = '-', color = 'blue', marker = 'v', markersize = 1.5)
ax.plot(T3_keys[::5]+T3_keys[-1:],T3_values[::5]+T3_values[-1:],linestyle = '-', color = 'green', marker = '^', markersize = 1.5)
ax.plot(T4_keys[::6]+T4_keys[-1:],T4_values[::6]+T4_values[-1:],linestyle = '-', color = 'teal', marker = 's', markersize = 1.5)
ax.plot(T5_mom_keys[::10]+T5_mom_keys[-1:],T5_mom_values[::10]+T5_mom_values[-1:],linestyle = '-', color = 'orange', marker = 'o', markersize = 1.5)
ax.plot(T5_keys[::10]+T5_keys[-1:],T5_values[::10]+T5_values[-1:],linestyle = '-', color = 'black', marker = 'D', markersize = 1.5)
ax.axhline(y = 10.25, linestyle = '--')
ax.legend(['$T$ = 1 MILD','$T$ = 2 MILD','$T$ = 3 MILD','$T$ = 4 MILD', '$T$ = 5 MILD','$T$ = 5 ALD'],loc = 'upper right',fontsize = 12)
ax.set_xlabel('NUMBER OF FUNCTION EVALUATIONS (NFEs)', fontsize = 12)
ax.set_ylabel('FRECHET INCEPTION DISTANCE (FID)', fontsize = 12)
ax.annotate('FID = 10.87',(0,15))
#ax.set_ylim([0,400])

plt.savefig("celeba_plot.pdf",format = 'pdf')
plt.show()
```
